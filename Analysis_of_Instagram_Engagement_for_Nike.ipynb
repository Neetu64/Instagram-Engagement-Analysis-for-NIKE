**Data Preparation**

```python
# Imports
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import confusion_matrix, accuracy_score
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer

# Download NLTK resources for text preprocessing
nltk.download('stopwords')
nltk.download('punkt')

# Initialize a stemmer and load English stop words
stemmer = SnowballStemmer("english")
stop_words = set(stopwords.words('english'))

# Function to process text by tokenizing, removing stop words, and stemming
def preprocess_text(text):
    tokens = word_tokenize(text.lower())  # Convert text to lowercase and tokenize it
    # Remove non-alphabetical words and stop words, then apply stemming
    filtered_tokens = [stemmer.stem(word) for word in tokens if word.isalpha() and word not in stop_words]
    return " ".join(filtered_tokens)  # Join the processed words back into a string

# Load the dataset containing Instagram post data
df = pd.read_csv('nike_instagram_posts_with_labels.csv')

# Create a binary column based on the median value of 'likes'
median_likes = df['likes'].median()  # Calculate the median likes
# Assign 1 if the number of likes is greater than or equal to the median, otherwise 0
df['binary'] = (df['likes'] >= median_likes).astype(int)

# Process and clean 'labels' and 'post_caption' columns
df['labels_text'] = df['labels'].apply(lambda x: ' '.join(eval(x)))
df['post_caption'] = df['post_caption'].fillna('')

# Apply the text preprocessing function to the 'labels_text' and 'post_caption' columns
df['post_caption'] = df['post_caption'].apply(preprocess_text)
df['labels_text'] = df['labels_text'].apply(preprocess_text)
```

**Modeling**

```python
# Vectorize the text columns using TF-IDF (Term Frequency-Inverse Document Frequency)

# for 'labels_text' column
vectorizer_labels = TfidfVectorizer(stop_words='english')
X_labels = vectorizer_labels.fit_transform(df['labels_text'])

# for 'post_caption' column
vectorizer_caption = TfidfVectorizer(stop_words='english')
X_caption = vectorizer_caption.fit_transform(df['post_caption'])

# Combine 'labels_text' and 'post_caption' columns into a single feature for combined analysis
combined_text = df['labels_text'] + ' ' + df['post_caption']
X_combined = vectorizer_caption.fit_transform(combined_text)

# Split the data into training and testing sets for each feature set (80% training, 20% testing)
X_train_labels, X_test_labels, y_train, y_test = train_test_split(X_labels, df['binary'], test_size=0.2, random_state=42)
X_train_caption, X_test_caption, _, _ = train_test_split(X_caption, df['binary'], test_size=0.2, random_state=42)
X_train_combined, X_test_combined, _, _ = train_test_split(X_combined, df['binary'], test_size=0.2, random_state=42)

# Initialize a Logistic Regression model
logreg = LogisticRegression(max_iter=1000)

# Train and evaluate the model using 'labels' only
logreg.fit(X_train_labels, y_train)
y_pred_labels = logreg.predict(X_test_labels)

# Train and evaluate the model using 'post_caption' only
logreg.fit(X_train_caption, y_train)
y_pred_caption = logreg.predict(X_test_caption)

# Train and evaluate the model using combined 'labels' and 'post_caption'
logreg.fit(X_train_combined, y_train)
y_pred_combined = logreg.predict(X_test_combined)
```

**Evaluation**

```python
# Function to evaluate the model using confusion matrix and accuracy score
def evaluate_model(y_test, y_pred, label):
    conf_matrix = confusion_matrix(y_test, y_pred)  # Calculate confusion matrix
    accuracy = accuracy_score(y_test, y_pred)  # Calculate accuracy score
    print(f"Results using {label}:")
    print("Confusion Matrix:")
    print(conf_matrix)
    print("Accuracy:", accuracy)

# Evaluate each case (using 'labels' only, 'post_caption' only, and both combined)
evaluate_model(y_test, y_pred_labels, "'labels' only")
evaluate_model(y_test, y_pred_caption, "'post_caption' only")
evaluate_model(y_test, y_pred_combined, "combined 'labels' and 'post_caption'")
```

```python
# Import necessary libraries
import os
import csv
import nltk
import lda
import pandas as pd
import numpy as np
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
# Download required NLTK resources if not already available
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python

# Load the data from an Excel file into a DataFrame
df = pd.read_csv('nike_instagram_posts_with_labels.csv').set_index('Unnamed: 0')
# Check for null values in the DataFrame and print the count of empty columns
df.isnull().sum() #it looks like we have a null post_caption
```

```python
df[df['post_caption'].isna()] #this is the row, lets drop it
```

```python
# Drop rows with any null values
df = df.dropna()

#reset index
df = df.reset_index()
df = df.drop(columns='index')
df.index.names = ['index']
```

```python
# Initialize tokenizer and lemmatizer
tokenizer = RegexpTokenizer(r'\w+')
lemmatizer = WordNetLemmatizer()

# Set of English stopwords
stop_words = set(stopwords.words('english'))


#Function to tokenize
# Tokenizes text by words
# Lemmatizes words to their base form
def preprocess_txt(text): #takes text column as an input
  #tokenize
  tokens = tokenizer.tokenize(text.lower())
  #take out stop words
  tokens = [word for word in tokens if word not in stop_words]
  #lemmatize
  tokens = [lemmatizer.lemmatize(word) for word in tokens]
  return tokens #return

# Create a CountVectorizer to transform text data into a matrix of token counts
vectorizer = CountVectorizer(analyzer=preprocess_txt)

# Fit and transform the text
X = vectorizer.fit_transform(df['text'])
```

```python
# Print the shape of the feature matrix
X.shape
```

```python
input_topics = 3
# Initialize and fit the LDA model
model = lda.LDA(n_iter=1000,n_topics=input_topics)
#fit
model.fit(X)
```

```python
#  Extract the topic-word distributio
topic_word_dist = model.topic_word_

# Get the vocabulary from the CountVectorizer
vocab = vectorizer.get_feature_names_out()

# Extract words for each topic
for k in range(input_topics): #for each topic
  topic_0_words = np.argsort(topic_word_dist[k])[::-1][:10]  # Get top 10 words
  topic_0_words_str = [vocab[i] for i in topic_0_words]

  print(f"Topic {k+1} words: {topic_0_words_str}")
```

We have come to around 3 topics using our dataset.

- Topic 1: athletics/athletes
- Topic 2: shoes
- Topic 3: artistic photography (graphics)


```python
# Extract document-topic distribution from the model
doc_topic_dist = model.doc_topic_

# Convert the document-topic distribution to a DataFrame
doc_topic_df = pd.DataFrame(doc_topic_dist, columns=['athletics', 'shoes', 'photography']).reset_index()

df = df.join(doc_topic_df, on='index',lsuffix='idx')
```

```python
df[:5]
```

```python
#sort from high to low
df = df.sort_values(by='likes',ascending=False)

#get quartiles
q1 = df['likes'].quantile(0.25)
q3 = df['likes'].quantile(0.75)

#get posts
df_high_likes = df[df['likes'] >= q3]
df_low_likes = df[df['likes'] <= q1]
```

```python
#get averages
avg_topics_high = df_high_likes[['athletics', 'shoes', 'photography']].mean(axis=0)
avg_topics_low = df_low_likes[['athletics', 'shoes', 'photography']].mean(axis=0)

#make a table to show
comparison_table = pd.DataFrame({'Quartile': ['High Likes', 'Low Likes'], #shoes quartile
                                  'Athletics': [avg_topics_high['athletics'], avg_topics_low['athletics']], #athletics mean
                                  'Shoes': [avg_topics_high['shoes'], avg_topics_low['shoes']], #shoes mean
                                  'Photography': [avg_topics_high['photography'], avg_topics_low['photography']]}) #photography mean

comparison_table
```
